## Config

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
from nb_util import data_path, device

model_name = 'celeba-cef-joint'
checkpoint_path = Path('/data/cef-models/')
gen_path = data_path / 'generated' / model_name
```

## Data

```python
import torchvision
from torch.utils.data import DataLoader
from torchvision import transforms
import data

channels = 3
height = 64
width = 64

transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
])

train_data = data.CelebA(root=data_path, split='train', transform=transform)
test_data = data.CelebA(root=data_path, split='test', transform=transform)
```

## Model

```python
import torch
from nflows import transforms, distributions, flows

n = height * width * channels
m = 1536

# Define a conformal embedding
conf_embedding = transforms.CompositeTransform([
    transforms.ConformalScaleShift(channels, m),
    transforms.HouseholderConv(channels, kernel_size=4),
    
    transforms.ConformalScaleShift(channels*16, m),
    transforms.ConditionalOrthogonal1x1Conv(channels*16, channels*8),
    
    transforms.ConformalScaleShift(channels*8, m),
    transforms.HouseholderConv(channels*8, kernel_size=4),
    
    transforms.ConformalScaleShift(channels*128, m),
    transforms.ConditionalOrthogonal1x1Conv(channels*128, channels*64),
    
    transforms.ConformalScaleShift(channels*64, m),
    transforms.HouseholderConv(channels*64, kernel_size=2),
    
    transforms.ConformalScaleShift(channels*256, m),
    transforms.Orthogonal1x1Conv(channels*256, channels*128),
    
    transforms.SqueezeTransform(factor=2),
    transforms.Flatten((m, 1, 1))
])

# Define an invertible backbone to learn the base distribution
backbone = (
    #flows.SimpleRealNVP(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))
    #flows.SimpleGlow(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))
    flows.SimpleNSF(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))

# Combine into a flow
flow = flows.Flow(transform=conf_embedding, distribution=backbone)
flow = flow.to(device)
```

## Train

Schedule training

```python
import torch.optim as opt

optim = opt.Adam(flow.parameters(), lr=0.0001)
scheduler = opt.lr_scheduler.MultiStepLR(optim, milestones=[10, 500, 1000], gamma=0.1)

def schedule():
    '''Yield epoch weights for likelihood and recon loss, respectively'''
    for _ in range(10):
        yield 0.00001, 10000
        scheduler.step()
    while True:
        yield 0.0001, 10000
        scheduler.step()
```

Create dataloaders

```python
train_batch_size = 128
test_batch_size = 128

train_loader = DataLoader(train_data, batch_size=train_batch_size, shuffle=True, num_workers=30)
test_loader = DataLoader(test_data, batch_size=test_batch_size, shuffle=True, num_workers=30)
```

If resuming from a checkpoint, load state from the specified epoch

```python
checkpoint_epoch = -1

if checkpoint_epoch >= 0:
    checkpoint = torch.load(checkpoint_path / f'{model_name}-e{checkpoint_epoch}.pt')
    
    flow.load_state_dict(checkpoint['model'])
    optim.load_state_dict(checkpoint['optimizer'])
    scheduler = checkpoint['scheduler']
```

Set up training visualizations

```python
num_recons = 4
num_samples = 8
sample_x = next(iter(train_loader))[0].to(device)[:num_recons]

with torch.no_grad():
    gen_samples = flow.sample(num_samples)
    sample_mid_latent, _ = conf_embedding.forward(sample_x)
    sample_recons, _ =  conf_embedding.inverse(sample_mid_latent)
```

Train the flow

```python
import numpy as np
import torch
import torch.nn as nn
from tqdm import tqdm
from nb_util import compare_batches, display_batch, update_displayed_batch

fig1, display_id1 = display_batch(gen_samples)
fig2, display_id2 = compare_batches(sample_x, sample_recons)

for epoch, (alpha, beta) in enumerate(schedule(), checkpoint_epoch+1):
    
    # Train for one epoch
    flow.train()
    progress_bar = tqdm(enumerate(train_loader))
    
    for batch, (image, _) in progress_bar:
        image = image.to(device)
        optim.zero_grad()

        # Compute reconstruction error
        with torch.set_grad_enabled(beta > 0):
            mid_latent, _ = conf_embedding.forward(image)
            reconstruction, log_conf_det = conf_embedding.inverse(mid_latent)
            reconstruction_error = torch.mean((image - reconstruction)**2)

        # Compute log likelihood
        with torch.set_grad_enabled(alpha > 0):
            log_pu = backbone.log_prob(mid_latent)
            log_likelihood = torch.mean(log_pu + log_conf_det)

        # Training step
        loss = - alpha*log_likelihood + beta*reconstruction_error
        loss.backward()
        optim.step()

        # Display results
        progress_bar.set_description(f'[E{epoch} B{batch}] | loss: {loss: 6.2f} | LL: {log_likelihood:6.2f} '
                                     f'| recon: {reconstruction_error:6.5f} ')
        if batch % 10 == 0:
            with torch.no_grad():
                gen_samples = flow.sample(num_samples)
                sample_mid_latent, _ = conf_embedding.forward(sample_x)
                sample_recons, _ =  conf_embedding.inverse(sample_mid_latent)
            update_displayed_batch(gen_samples, fig1, display_id1)
            compare_batches(sample_x, sample_recons, fig2, display_id2)
    
    # Evaluate on the test set
    flow.eval()
    recon_errors = []
    log_likelihoods = []
    
    for image, _ in test_loader:
        image = image.to(device)
        
        with torch.no_grad():
            # Compute reconstruction error
            mid_latent, _ = conf_embedding.forward(image)
            reconstruction, log_conf_det = conf_embedding.inverse(mid_latent)
            reconstruction_error = torch.mean((image - reconstruction)**2, dim=(1, 2, 3))
            recon_errors.extend(reconstruction_error.tolist())

            # Compute log likelihood
            log_pu = backbone.log_prob(mid_latent)
            log_likelihood = log_pu + log_conf_det
            log_likelihoods.extend(log_likelihood.tolist())
    
    test_recon_error = np.mean(recon_errors)
    test_log_likelihood = np.mean(log_likelihoods)
    test_loss = -alpha*test_log_likelihood + beta*test_recon_error
    print(f'[E{epoch} val] | loss: {test_loss} | LL: {test_log_likelihood} | test recon: {test_recon_error}')
        
    # Save a checkpoint
    checkpoint = {
        'epoch': epoch,
        'batch_size': train_batch_size,
        'alpha': alpha,
        'beta': beta,
        'reconstruction_error': test_recon_error,
        'log_likelihood': test_log_likelihood,
        'model': flow.state_dict(),
        'optimizer': optim.state_dict(),
        'scheduler': scheduler,
    }        
    torch.save(checkpoint, checkpoint_path / f'{model_name}-e{epoch}.pt')
```
