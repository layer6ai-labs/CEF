## Config

```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
from nb_util import data_path, device

model_name = 'mnist-cef-joint'
checkpoint_path = Path('/data/cef-models/')
```

## Data

```python
import torchvision
from torch.utils.data import DataLoader
from torchvision import transforms

# Pad images from 28x28 to 32x32 to make it a power of 2
channels = 1
height = 32
width = 32

transform = transforms.Compose([
    transforms.Pad(2),
    transforms.ToTensor(),
])

train_data = torchvision.datasets.MNIST(
    root=data_path, train=True, download=True, transform=transform)
test_data = torchvision.datasets.MNIST(
    root=data_path, train=False, download=True, transform=transform)
```

## Model

```python
import torch
from nflows import transforms, distributions, flows

n = height * width * channels
m = 128

# Define a conformal embedding
conf_embedding = transforms.CompositeTransform([
    transforms.ConformalScaleShift(channels, m),
    transforms.HouseholderConv(channels, kernel_size=8),
    
    transforms.ConformalScaleShift(channels*64, m),
    transforms.ConditionalOrthogonal1x1Conv(channels*64),
    
    transforms.ConformalScaleShift(channels*64, m),
    transforms.SqueezeTransform(factor=height//8),
    
    transforms.Orthogonal1x1Conv(channels*1024, m),
    
    transforms.Flatten((m, 1, 1))
])

# Define an invertible backbone to learn the base distribution
backbone = (
    #flows.SimpleRealNVP(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))
    #flows.SimpleGlow(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))
    flows.SimpleNSF(features=m, hidden_features=512, num_layers=8, num_blocks_per_layer=3))

# Combine into a flow
flow = flows.Flow(transform=conf_embedding, distribution=backbone)
flow = flow.to(device)
```

## Train

Schedule optimizer

```python
import torch.optim as opt

optim = opt.Adam(flow.parameters(), lr=0.001)
scheduler = opt.lr_scheduler.MultiStepLR(optim, milestones=[250], gamma=0.1)

def schedule():
    '''Yield epoch weights for likelihood and recon loss, respectively'''
    while True:
        yield 0.01, 100000 
        scheduler.step()
```

Create dataloaders

```python
train_batch_size = 32
test_batch_size = 32

train_loader = DataLoader(train_data, batch_size=train_batch_size, shuffle=True, num_workers=30)
test_loader = DataLoader(test_data, batch_size=test_batch_size, shuffle=True, num_workers=30)
```

If resuming from a checkpoint, load state from the specified epoch

```python
log_likelihoods = []

for checkpoint_epoch in range(1200):
    checkpoint = torch.load(checkpoint_path / f'{model_name}-e{checkpoint_epoch}.pt', 
                            map_location=torch.device('cpu'))
    log_likelihoods.append(checkpoint['log_likelihood'])
    print(f'LL {log_likelihoods[-1]}')
```

```python
import matplotlib.pyplot as plt

plt.plot(log_likelihoods[550:])
```

```python
import numpy as np

np.argmax(log_likelihoods)
```

```python
checkpoint_epoch = 3000

if checkpoint_epoch >= 0:
    checkpoint = torch.load(checkpoint_path / f'{model_name}-e{checkpoint_epoch}.pt')
    
    flow.load_state_dict(checkpoint['model'])
    optim.load_state_dict(checkpoint['optimizer'])
    scheduler = checkpoint['scheduler']
```

```python
num_recons = 4
num_samples = 32
temp = 1
sample_x = next(iter(train_loader))[0].to(device)[:num_recons]

with torch.no_grad():
    sample_mid_latent, _ = conf_embedding.forward(sample_x)
    sample_recons, _ =  conf_embedding.inverse(sample_mid_latent)
        
    latent_samples = torch.normal(mean=torch.zeros(num_samples, m), std=torch.ones(num_samples, m)*temp)
    mid_latent_samples, _ = backbone._transform.inverse(latent_samples.to(device))
    gen_samples, _ = conf_embedding.inverse(mid_latent_samples)
```
